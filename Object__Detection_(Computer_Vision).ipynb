# **Step 2: Implement Object Detection**

---



**2.1 Load YOLOv8 Model**

```python
from ultralytics import YOLO

# Load pretrained YOLOv8 model (medium size)
model = YOLO('yolov8m.pt')  # Options: yolov8n.pt (nano), yolov8s.pt (small), yolov8m.pt (medium), yolov8l.pt (large)
```

**2.2 Detect Objects in an Image**

```python
import cv2
from matplotlib import pyplot as plt
import zipfile
import os

def detect_objects(image_path):
    # Perform detection
    results = model(image_path)

    # Show results
    for r in results:
        im_array = r.plot()  # Plot detection results
        im_rgb = cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)
        plt.figure(figsize=(12, 8))
        plt.imshow(im_rgb)
        plt.axis('off')
        plt.show()

    return results

# Example usage (upload your image first)
from google.colab import files
uploaded = files.upload()
zip_path = list(uploaded.keys())[0]

# Extract the zip file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("uploaded_images")

# Process each image in the extracted directory
image_files = [f for f in os.listdir("uploaded_images") if f.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif', '.webp', '.heic'))]

if not image_files:
    print("No supported image files found in the zip archive.")
else:
    for image_file in image_files:
        image_path = os.path.join("uploaded_images", image_file)
        print(f"Processing {image_file}...")
        detect_objects(image_path)
```

```python
# 1. Install packages
!pip install ultralytics flask pyngrok

# 2. Imports
from ultralytics import YOLO
from flask import Flask, request, jsonify
import cv2
import numpy as np
from pyngrok import ngrok

# 3. Initialize YOLO
model = YOLO('yolov8n.pt')

# 4. Flask app
app = Flask(__name__)

# 5. Set Ngrok token (paste your actual token here)
ngrok.set_auth_token("your_actual_ngrok_token_here")

# 6. Detection endpoint
@app.route('/detect', methods=['POST'])
def detect():
    if 'file' not in request.files:
        return jsonify({'error': 'No file uploaded'}), 400

    file = request.files['file'].read()
    img = cv2.imdecode(np.frombuffer(file, np.uint8), cv2.IMREAD_COLOR)

    results = model(img)
    detections = []

    for box in results[0].boxes:
        detections.append({
            'class': results[0].names[box.cls.item()],
            'confidence': float(box.conf.item()),
            'bbox': box.xyxy[0].tolist()
        })

    return jsonify(detections)

# 7. Start Ngrok (only if token is set)
try:
    public_url = ngrok.connect(5000).public_url
    print(f" * Running on {public_url}")
except Exception as e:
    print("Ngrok error:", str(e))
    print("Using Colab's built-in port forwarding instead:")
    from google.colab.output import eval_js
    print("Access at:", eval_js("google.colab.kernel.proxyPort(5000)"))

# 8. Run app
app.run()
```
